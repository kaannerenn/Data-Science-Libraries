```python
import numpy as np
import nltk
import random
nltk.download('words')
from nltk.corpus import words
import string
from IPython.display import display, clear_output
import ipywidgets as widgets
```

```python
# 10x10 boş bir numpy array oluştur.
# Karakterler için string dtype seç.
# Başlangıçta boşları “-” gibi bir placeholder ile doldur.
word_array = np.full((10, 10), '-', dtype='<U1')
```

```python
# Aslında while True ile yerleştirene kadar yapabiliriz. Fakat gerçekten yerleştirilemezse sonsuz döngüye girer.
# Sağa doğru yerleştiren fonksyion.
def put_word_right_side(word, array, max_attempts=100): 
    rows, cols = array.shape
    word_len = len(word)
    
    attempts = 0
    while attempts < max_attempts:
        attempts += 1
        
        
        i = np.random.randint(0, rows)
        j = np.random.randint(0, cols - word_len + 1)
        
        
        can_place = True
        for k in range(word_len):
            if array[i, j + k] != '-' and array[i, j + k] != word[k]:
                can_place = False
                break
        
        
        if can_place:
            for k in range(word_len):
                array[i, j + k] = word[k]
            return True
    
    
    return False
```

```python
# Aşağı doğru yerleştiren fonksiyon
def put_word_down_side(word,array, max_attempts = 100):
    rows, cols = array.shape
    word_len = len(word)
    
    attempts = 0
    while attempts < max_attempts:
        attempts += 1
        
        i = np.random.randint(0,rows - word_len + 1)
        j = np.random.randint(0,cols)
        
        can_place = True
        for k in range(word_len):
            if array[i + k, j] != '-' and array[i+k,j] != word[k]:
                can_place = False
                break
        if can_place:
            for k in range(word_len):
                array[i + k, j] = word[k]
            return True
    return False
```

```python
# Çapraz sağa yerleştiren fonksiyon
def put_diag_right_side(word,array,max_attempts = 100):
    rows, cols = array.shape
    word_len = len(word)
    
    attempts = 0
    
    while attempts < max_attempts:
        attempts+=1
        
        i = np.random.randint(0,rows - word_len + 1)
        j = np.random.randint(0,cols - word_len + 1)
        
        can_place = True
        for k in range(word_len):
            if array[i + k, j + k] != '-' and array[i + k, j + k] != word[k]:
                can_place = False
                break
        if can_place:
            for k in range(word_len):
                array[i + k, j + k] = word[k]
            return True 
    return False
```

```python
# Çapraz sola yerleştiren fonksiyon
def put_diag_left_side(word,array,max_attempts = 100):
    rows, cols = array.shape
    word_len = len(word)
    
    attempts = 0
    
    while attempts < max_attempts:
        attempts+=1
        
        i = np.random.randint(0,rows - word_len + 1)
        j = np.random.randint(word_len - 1, cols)
        
        can_place = True
        for k in range(word_len):
            if array[i + k, j - k] != '-' and array[i + k, j - k] != word[k]:
                can_place = False
                break
        if can_place:
            for k in range(word_len):
                array[i + k, j - k] = word[k]
            return True 
    return False
```

```python
def random_word(length):
    # Tüm İngilizce kelimeler
    word_list = words.words()
    
    filtered_words = [w for w in word_list if len(w) == length]
     
    if filtered_words:
        return random.choice(filtered_words)
    else:
        return None 

def choose_side(process_word,array):
    
    sides = [put_diag_left_side,put_diag_right_side,put_word_down_side,put_word_right_side]
    
    choice_side = random.choice(sides)
    
    return choice_side(process_word,array)
    
k = 1
while k <= 4:
    x = np.random.randint(3,8)
    word = random_word(x)
    print(word) # Seçilen kelimeyi görmek için koydum.
    choose_side(word,word_array)
    k += 1

print(word_array)
rows, cols = word_array.shape
for i in range(rows):
    for j in range(cols):
        if word_array[i,j] == '-':
            word_array[i,j] = random.choice(string.ascii_letters).lower()
print(word_array)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Grid gösterimi fonksiyonu
def display_grid(array, highlight_indices=None):
    clear_output(wait=True)
    for i in range(array.shape[0]):
        row_str = ""
        for j in range(array.shape[1]):
            char = array[i,j]
            if highlight_indices and (i,j) in highlight_indices:
                row_str += f"\033[91m{char}\033[0m "  # kırmızı renk
            else:
                row_str += f"{char} "
        print(row_str)
    print("\n")

# Kelime arama fonksiyonu
def find_word(word, array):
    word = word.lower()  
    indices = []
    rows, cols = array.shape
    word_len = len(word)

    # Sağa
    for i in range(rows):
        for j in range(cols - word_len + 1):
            if "".join(array[i,j:j+word_len]).lower() == word:
                return [(i, j+k) for k in range(word_len)]

    # Aşağı
    for i in range(rows - word_len + 1):
        for j in range(cols):
            if "".join(array[i+k,j].lower() for k in range(word_len)) == word:
                return [(i+k, j) for k in range(word_len)]

    # Çapraz sağa
    for i in range(rows - word_len + 1):
        for j in range(cols - word_len + 1):
            if "".join(array[i+k, j+k].lower() for k in range(word_len)) == word:
                return [(i+k, j+k) for k in range(word_len)]

    # Çapraz sola
    for i in range(rows - word_len + 1):
        for j in range(word_len - 1, cols):
            if "".join(array[i+k, j-k].lower() for k in range(word_len)) == word:
                return [(i+k, j-k) for k in range(word_len)]

    return None


word_input = widgets.Text(placeholder='Kelimeyi gir...')
check_button = widgets.Button(description="Kontrol Et")
output = widgets.Output()


def check_word(b):
    with output:
        indices = find_word(word_input.value, word_array)
        if indices:
            display_grid(word_array, highlight_indices=indices)
            print(f"Tebrik! '{word_input.value}' bulundu.")
        else:
            display_grid(word_array)
            print(f"Üzgünüm, '{word_input.value}' bulunamadı.")

check_button.on_click(check_word)

display(widgets.VBox([word_input, check_button, output]))
with output:
    display_grid(word_array)
```
